## Scaling data

```python
from sklearn.preprocessing import StandardScaler

# Initialize the StandardScaler
scaler = StandardScaler()

# Fit the scaler to the data and transform it
scaled_data = scaler.fit_transform(data)

# Convert the scaled data back to a DataFrame with the original column names
scaled_df = pd.DataFrame(scaled_data, columns=data.columns)

# Display the first few rows of the scaled DataFrame
scaled_df.head()
```

## K-Means Clustering

```python
from sklearn.cluster import KMeans

# Initialize list to store inertia values
inertia_values = []

# Range of cluster values to evaluate
K = range(1, 11)

# Iterate over each K value to calculate inertia (within-cluster sum of squares)
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)  # Added random_state for reproducibility
    kmeans.fit(data)
    inertia_values.append(kmeans.inertia_)

# Plot the Elbow curve
plt.figure(figsize=(8, 6))
plt.plot(K, inertia_values, marker='o', color='b', linestyle='--', markersize=8)
plt.title('Elbow Method for Optimal K', fontsize=15)
plt.xlabel('Number of Clusters', fontsize=12)
plt.ylabel('Inertia (Sum of Squared Distances)', fontsize=12)
plt.grid(True)
plt.show()
```

```python
from sklearn import metrics

# Initialize and fit the KMeans model with 3 clusters
kmeans = KMeans(n_clusters=3, random_state=111)
kmeans.fit(scaled_df)
```

```python
# Count the number of records in each cluster
cluster_counts = pd.Series(kmeans.labels_).value_counts()
print("Cluster counts:\n", cluster_counts)
```

```python
# Calculate the Silhouette Coefficient to evaluate the model's quality
silhouette_score = metrics.silhouette_score(scaled_df, kmeans.labels_)
print(f"Silhouette Coefficient (K=3): {silhouette_score:.4f}")
```

## Prediction

```python
# Predicting the cluster labels for each data point
cluster_labels = kmeans.fit_predict(scaled_df)

# Assign the predicted cluster labels to the original DataFrame
kmeans_df = pd.DataFrame(data.copy())  # Used df.copy() to avoid modifying the original DataFrame
kmeans_df['KMeans_Clusters'] = cluster_labels

# Display the first 10 rows of the DataFrame with the added cluster labels
kmeans_df.head(10)
```

## Visualization of clusters

```python
# Visualizing clusters using a scatter plot for Child Mortality vs. GDP per capita (gdpp)
plt.figure(figsize=(8, 6))  # Set figure size for better clarity
sns.scatterplot(x='child_mort', y='gdpp', hue='KMeans_Clusters', data=kmeans_df, palette='viridis', s=100, edgecolor='black')

# Adding title and labels to the plot
plt.title("Child Mortality vs GDP per Capita", fontsize=15)
plt.xlabel("Child Mortality", fontsize=12)
plt.ylabel("GDP per Capita", fontsize=12)

# Display the plot
plt.legend(title='Clusters', loc='upper right')  # Added legend for clarity
plt.show()
```

```python
# Visualizing clusters using a scatter plot for Inflation vs. GDP per capita (gdpp)
plt.figure(figsize=(8, 6))  # Set figure size for better clarity
sns.scatterplot(x='inflation', y='gdpp', hue='KMeans_Clusters', data=kmeans_df, palette='viridis', s=100, edgecolor='black')

# Adding title and labels to the plot
plt.title("Inflation vs GDP per Capita", fontsize=15)
plt.xlabel("Inflation", fontsize=12)
plt.ylabel("GDP per Capita", fontsize=12)

# Display the plot with a legend
plt.legend(title='Clusters', loc='upper right')
plt.show()
```

## Hierarchical Clustering

```python
import scipy.cluster.hierarchy as hcluster

# Set figure size for better readability
plt.figure(figsize=(15, 8))

# Perform hierarchical clustering and plot the dendrogram
dendrogram = hcluster.dendrogram(
    hcluster.linkage(scaled_df, method='ward')
)

# Adding title and labels to the plot
plt.title("Dendrogram for Hierarchical Clustering", fontsize=18)
plt.xlabel("Index", fontsize=14)
plt.ylabel("Distance", fontsize=14)

# Display the plot
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering

# Perform Agglomerative Hierarchical Clustering (no need for affinity argument)
hcluster = AgglomerativeClustering(n_clusters=3, linkage='ward')

# Predict the cluster labels
hcluster_labels = hcluster.fit_predict(scaled_df)

# Create a new DataFrame from the original data and add the cluster labels
hcluster_df = data.copy()  # Use df.copy() to avoid modifying the original DataFrame
hcluster_df['hcluster'] = hcluster_labels

# Display the first few rows of the DataFrame with the added labels
hcluster_df.head()
```

## Visualization Hierarchical Clustering

```python
# Visualizing the results of Hierarchical Clustering: Child Mortality vs GDPP
plt.figure(figsize=(10, 6))  # Set figure size for better clarity

# Scatter plot with hue based on hierarchical clustering labels
sns.scatterplot(data=hcluster_df, x='child_mort', y='gdpp', hue='hcluster', palette='viridis', s=100)

# Adding title and labels to the plot for clarity
plt.title("Child Mortality vs GDPP (Hierarchical Clustering)", fontsize=16)
plt.xlabel("Child Mortality", fontsize=14)
plt.ylabel("GDP per Capita (GDPP)", fontsize=14)

# Display the plot
plt.legend(title="Cluster", fontsize=12)  # Add legend with custom title
plt.show()
```

```python
# Visualizing the results of Hierarchical Clustering: Inflation vs GDPP
plt.figure(figsize=(10, 6))  # Set figure size for better clarity

# Scatter plot with hue based on hierarchical clustering labels
sns.scatterplot(data=hcluster_df, x='inflation', y='gdpp', hue='hcluster', palette='coolwarm', s=100)

# Adding title and labels to the plot for better clarity
plt.title("Inflation vs GDP per Capita (GDPP)", fontsize=16)
plt.xlabel("Inflation (%)", fontsize=14)
plt.ylabel("GDP per Capita (GDPP)", fontsize=14)

# Display the plot
plt.legend(title="Cluster", fontsize=12)  # Customizing legend
plt.show()
```
