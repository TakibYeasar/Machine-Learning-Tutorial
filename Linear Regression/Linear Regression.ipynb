```python
# make function for count plot and scatter plots for Categorical features

def Categorical_Plot(column, rotation=0):
    plt.figure(figsize=(30, 10))

    # Count Plot (with colors)
    plt.subplot(1, 2, 1)
    colors = sns.color_palette("viridis", len(pd.value_counts(data[column])))  # Color scheme
    pd.value_counts(data[column]).plot(kind='bar', color=colors)
    
    plt.xticks(horizontalalignment="center", fontsize=15, rotation=rotation)
    plt.xlabel(f"\n{str(column)}", fontsize=15, color="darkblue")
    plt.ylabel("Count\n", fontsize=15, color="darkblue")
    plt.yticks(fontsize=15, color="darkred")
    plt.title(f"{str(column)} - Count\n", fontsize=15, color="purple")
    
    # Scatter Plot (with colors)
    plt.subplot(1, 2, 2)
    sns.barplot(x=data[column], y=data["price"], palette="coolwarm")  # Adding color gradient
    
    plt.xticks(horizontalalignment="center", fontsize=15, rotation=rotation, color="darkred")
    plt.xlabel(f"\n{str(column)}", fontsize=15, color="darkblue")
    plt.yticks(fontsize=15, color="darkred")
    plt.title(f"{str(column)} - Price\n", fontsize=15, color="purple")
    plt.ylabel("Price\n", fontsize=15, color="darkblue")

    plt.show()
    print()
```

```python
# make function for box plots and scatter plots for Numerical features

def Numerical_Plot(column):
    plt.figure(figsize=(20, 5))

    # Box Plot with Colors
    plt.subplot(1, 2, 1)
    plt.grid(color='gray', linestyle='--', linewidth=0.5)  # Light grid for better clarity
    sns.boxplot(data[column], color="lightcoral")  # Box plot with color
    plt.xticks(horizontalalignment="center", fontsize=15, color="darkblue")
    plt.xlabel(f"\n{str(column)}", fontsize=15, color="darkblue")
    plt.yticks(fontsize=15, color="darkred")
    plt.title(f"\n{str(column)} Box Plot\n", fontsize=15, color="purple")

    # Scatter Plot with Colors
    plt.subplot(1, 2, 2)
    plt.grid(color='gray', linestyle='--', linewidth=0.5)  # Light grid for better clarity
    sns.scatterplot(x=data[column], y=data["price"], hue=data[column], palette="coolwarm")  # Colorful scatter plot
    plt.xticks(horizontalalignment="center", fontsize=15, color="darkred")
    plt.xlabel(f"\n{str(column)}", fontsize=15, color="darkblue")
    plt.yticks(fontsize=15, color="darkred")
    plt.title(f"{str(column)} - Price\n", fontsize=15, color="purple")
    plt.ylabel("Price\n", fontsize=15, color="darkblue")

    plt.show()
    print()
```

```python
house = pd.get_dummies(data, columns=['city'], prefix = ['city'])
```

```python
house = house.drop(['date', 'street', 'statezip', 'country','sqft_above'], axis = 1)
house
```

## Normalize

```python
columns = house.columns
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn import preprocessing

scaler = preprocessing.MinMaxScaler(feature_range = (0, 1))
normal = pd.DataFrame(scaler.fit_transform(house), columns = columns)
normal.head(20)
```

```python
house.shape
```

## Modeling

```python
x = normal.drop("price", axis=1)
y = pd.DataFrame(normal["price"])
```

```python
from sklearn.neighbors import NearestNeighbors

nbrs = NearestNeighbors(n_neighbors=3).fit(x)
neigh_dist, neigh_ind = nbrs.kneighbors(x)
sort_neigh_dist = np.sort(neigh_dist, axis=0)
```

```python
k_dist = sort_neigh_dist[:, 2]
plt.plot(k_dist)
plt.ylabel("k-NN distance")
plt.xlabel("Sorted observations (4th NN)")
plt.show()
```

```python
from sklearn.cluster import DBSCAN

clusters = DBSCAN(eps=0.9, min_samples=3).fit(x)
clusters.labels_
```

```python
x["noise"] = clusters.labels_
y["noise"] = clusters.labels_
```

```python
x = x[x.noise>-1]
y = y[y.noise>-1]
x.drop('noise', inplace = True, axis=1)
y.drop('noise', inplace = True, axis=1)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size = 0.2, random_state =43 )
```

```python
print(X_train.shape)
print(Y_train.shape)
print(X_test.shape)
print(Y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, Y_train)

Y_pred = model.predict(X_test)
```

```python
model.intercept_
```

```python
model.coef_
```

## Model Evaluations

```python
from sklearn import metrics

# MAE
print("Mean Absolute Error     : ", metrics.mean_absolute_error(Y_test, Y_pred))

# MSE
print("Mean Squared Error      : ", metrics.mean_squared_error(Y_test, Y_pred))

# RMSE
print("Root Mean Squared Error : ", np.sqrt(metrics.mean_squared_error(Y_test, Y_pred)))

#R2 Score
print("R2 Score                : ", metrics.r2_score(Y_test, Y_pred))
```
