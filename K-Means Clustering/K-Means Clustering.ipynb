```python
import plotly.express as px

# Create a pie chart using the correct column name
fig = px.pie(
    data, 
    names="Gender",
    color="Gender", 
    title="Gender Distribution", 
    color_discrete_sequence=["#0044FF", "#DE00FF"]
)

# Update trace properties to enhance visualization
fig.update_traces(
    textinfo="value",
    textfont=dict(size=20),
    marker=dict(
        line=dict(color="#FFFFFF", width=2)
    )
)

# Update layout for better aesthetics
fig.update_layout(
    font=dict(family="PT Sans", size=20),
    title=dict(
        text="Gender Distribution",
        font=dict(size=24),
        x=0.5
    )
)

# Show the figure
fig.show()
```

```python
data_new = data.drop(columns="Gender")
```

```python
import plotly.express as px
from sklearn.cluster import KMeans

# Define the color scale dynamically based on number of clusters
def get_color_scale(n_clusters):
    return px.colors.qualitative.Set1[:n_clusters]

# Loop through different cluster counts
for clusters in range(2, 11):    
    kms = KMeans(n_clusters=clusters, init="k-means++", random_state=42)
    kcluster = pd.DataFrame(kms.fit_predict(data_new), columns=["kcluster"])
    data_new["cluster"] = kcluster

    # Generate 3D scatter plot with dynamically updated color scale
    fig = px.scatter_3d(
        data_new, 
        x="Spending Score (1-100)", 
        y="Annual Income (k$)", 
        z="Age",
        hover_data=["Age"], 
        color="cluster", 
        opacity=0.8,
        color_continuous_scale=get_color_scale(clusters),
        title=f"K-Means Clustering Mall Customers  (Clusters = {clusters})\nSpending Score, Annual Income, Age"
    )

    # Update trace properties for better visualization
    fig.update_traces(
        marker=dict(
            size=5, 
            line=dict(width=2, color="white")
        )
    )
    
    # Hide color scale
    fig.update_coloraxes(showscale=False)

    # Update layout with better fonts and margins for better readability
    fig.update_layout(
        font=dict(family="PT Sans", size=14, color="#222222"),
        scene=dict(
            xaxis_title="Spending Score (1-100)", 
            yaxis_title="Annual Income (k$)", 
            zaxis_title="Age"
        ),
        margin=dict(l=20, r=20, t=40, b=20)  # Adjusting margins for clarity
    )

    # Show the plot
    fig.show()
```
